```python
import pandas as pd

df = pd.read_csv("/workspaces/data_3500_in_class_code/Lecture_code/titanic.csv")
df.head()
```

Programming Activity 1
Using the Titanic dataset and the Pandas library, fill in any missing data cells in the 'ages' column with the median age.

```python
df = df.rename(columns={"Pclass":"Class"})
df = df.rename(columns={"Parch":"Parent/Child"})
df = df.rename(columns={"SibSp":"Sibling/Spouse"})
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
df.isnull().sum()
```

Programming Activity 2
Using the Titanic dataset and the Pandas library, create a new column called 'FamilySize' and fill it with the size of the passenger's family.

```python
df["Family Size"] = df["Sibling/Spouse"] + df["Parent/Child"] + 1
df.head()
```

Programming Activity 3

Using the Titanic dataset and the Pandas library, find the average survival rate by class and sex.

```python
df["Survived"] = df["Survived"].astype(bool)
df.groupby(["Class", "Sex"])["Survived"].mean()
```

Programming Activity 4

Using the Titanic dataset and the Pandas library, find the average fare per class.

```python
df.groupby("Class")["Fare"].mean()
```
