```python
import pandas as pd

df = pd.read_csv("/workspaces/data_3500_in_class_code/Lecture_code/titanic.csv")
df.head()
```

Programming Activity 1
Using the Titanic dataset and the Pandas library, fill in any missing data cells in the 'ages' column with the median age.

```python
df = df.rename(columns={"Pclass":"Class"})
df = df.rename(columns={"Parch":"Parent/Child"})
df = df.rename(columns={"SibSp":"Sibling/Spouse"})
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
```

```python
df.isnull().sum()
```

Programming Activity 2
Using the Titanic dataset and the Pandas library, create a new column called 'FamilySize' and fill it with the size of the passenger's family.

```python
df["Family Size"] = df["Sibling/Spouse"] + df["Parent/Child"] + 1
df.head()
```

Programming Activity 3

Using the Titanic dataset and the Pandas library, find the average survival rate by class and sex.

```python
df["Survived"] = df["Survived"].astype(bool)
df.groupby(["Class", "Sex"])["Survived"].mean()
```

Programming Activity 4

Using the Titanic dataset and the Pandas library, find the average fare per class.

```python
average_fare_by_class = df.groupby("Class")["Fare"].mean()
print("Average Fare by Class:")
print(average_fare_by_class)
```

Programming Activity 5

Using the Titanic dataset and the Pandas library, find the median age of survivors versus non-survivors.

```python
median_age_by_survival = df.groupby("Survived")["Age"].median()
print("Median age by survival status (False = did not survive, True = survived):")
print(median_age_by_survival)
```

```python
# Weather dataframe

df2 = pd.read_csv("/workspaces/data_3500_in_class_code/Lecture_code/SLC_weather.csv")
df2.head()
```

Programming Activity 6

Using the weather file (located on week 11 programming activities page), find the longest warm streak where the high temperature was above 65 degrees.

```python
df2["Warm"] = df2["Max Temp"] > 65

current_streak = 0
longest_streak = 0

for is_warm in df2["Warm"]:
    if is_warm:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
    else:
        current_streak = 0  # reset when itâ€™s not a hot day

print("Longest hot streak:", longest_streak, "days")
```

Programming Activity 7

Using the same weather dataset, find the hottest day and the coldest night.

```python
coldest_night = df2.loc[df2["Min Temp"].idxmin()]
warmest_day = df2.loc[df2["Max Temp"].idxmax()]

print("Coldest night:")
print(coldest_night[["Date", "Min Temp"]])
print("\nWarmest day:")
print(warmest_day[["Date", "Max Temp"]])
```

Programming Activity 8

Using the same weather dataset, find the single most volatile day (the day with the biggest difference between high and low temperature).

```python
df2["Volatility"] = df2["Max Temp"] - df2["Min Temp"]

# Find the row with the highest volatility
most_volatile_day = df2.loc[df2["Volatility"].idxmax()]

print("Most volatile day:")
print(most_volatile_day[["Date", "Max Temp", "Min Temp", "Volatility"]])
```
